```python
from datasets import Dataset
from ragas import evaluate
from ragas.metrics import faithfulness, answer_relevancy, context_precision, context_recall
from ragas.llms import LangchainLLMWrapper
from langchain_openai import ChatOpenAI

```

```python
import os
from dotenv import load_dotenv

load_dotenv()
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
```

```python
questions = [
    "When was the first super bowl?",
    "What were some important historical events that happened in Boston in 1919?",
    "Find drawings included on grave markers from people who died in the late 1700s.",
    "In which year did the Boston Marathon bombings occur?",
    "Who has the concert called \"Eras Tour,\" and how many shows has she held?",
    "What painting contains a girl who smiles, and from every angle she seems to be smiling at you?"
]

contexts = [
    ["The first Super Bowl followed the 1966 season, replacing the NFL Championship Game. It was held on January 15, 1967."],
    ["In 1919, Boston experienced the Molasses Disaster, a major police strike, and the Victory Parade of the 101st Infantry Regiment."],
    ["In memory of Capt. Thomas Eaton; Loae Parker; Mr. Nathaniel Bachelor."],
    [
        "The Boston Marathon bombings occurred on April 15, 2013, near the finish line of the annual marathon.",
        "The attack killed three people and injured hundreds, leading to a large-scale manhunt across the Boston area."
    ],
    [
        "Taylor Swift's Eras Tour is a global concert tour that began in March 2023 and spans multiple continents.",
        "As of mid-2024, the Eras Tour has more than 150 scheduled shows across North America, South America, Europe, Asia, and Oceania.",
        "The tour features performances of songs from all ten of Swift’s studio albums and has broken numerous box-office records."
    ],
    [
        "The Mona Lisa is a portrait painting by Leonardo da Vinci, created between 1503 and 1506.",
        "It is famous for the subject’s enigmatic smile, which appears to change depending on the viewer’s angle.",
        "The painting depicts Lisa Gherardini, the wife of Francesco del Giocondo, and is displayed in the Louvre Museum in Paris."
    ]
]

answers = [
    "The first superbowl was held on January 15, 1967",
    "In 1919, Boston saw the Molasses Disaster, a major police strike, and the Victory Parade of the 101st Infantry Regiment.",
    "There are gravestone rubbings with drawings from the late 1700s, such as those of Capt. Thomas Eaton.",
    "The Boston Marathon bombings took place in 2013.",
    "The Eras Tour is by Taylor Swift, and she has performed more than 150 shows around the world since 2023.",
    "The painting you are describing is the Mona Lisa by Leonardo da Vinci."
]

ground_truths = [
    ["The first superbowl was held on January 15, 1967."],
    ["Boston’s key 1919 events included the Molasses Disaster, the police strike, and the post-war Victory Parade."],
    ["The collection includes gravestone drawings and rubbings from the 18th century, including Capt. Eaton and Loae Parker."],
    ["The Boston Marathon bombings happened in 2013."],
    ["Taylor Swift is the artist who has the concert called the Eras Tour, which includes about 150 shows worldwide between 2023 and 2024."],
    ["The painting that shows a woman who seems to smile from every angle is the Mona Lisa, painted by Leonardo da Vinci between 1503 and 1506."]
]
references = [" ".join(gts) for gts in ground_truths]
```

```python
dataset = Dataset.from_dict({
    "question": questions,
    "contexts": contexts,
    "answer": answers,
    "reference": references
})
```

```python
llm = LangchainLLMWrapper(ChatOpenAI(model="gpt-4o-mini", temperature=0))
```

```python
result = evaluate(
    dataset=dataset,
    metrics=[faithfulness, answer_relevancy, context_precision, context_recall],
    llm=llm,
)
result
```
